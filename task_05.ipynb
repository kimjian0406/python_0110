[math 모듈]

- 수학과 관련된 기능을 가지고 있음

math 모듈 내 함수
- sin(x) : 사인 값을 구함
- cos(x) : 코사인 값을 구함
- tan(x) : 탄젠트 값을 구함
- ceil(x) : 올림
- floor(x) : 내림

이 외에도 굉장히 많음....

````python
import math

print(math.sin(2))
print(math.cos(3))
print(math.ceil(4.7))
print(math.floor(6.8))
````

[from 구문]

모듈안에 수 많은 함수 중 우리가 원하는 함수만을 가져와 사용하면 문법적으로 굉장히 편해집니다.

```
from 모듈 이름 import 가져오고 싶은 변수 또는 함수
```

----
from math import sin

(기존)
math.sin(5)

(from 구문 사용 후)
sin(5)

````python
from math import sin, cos, ceil, floor

print(sin(1))
print(cos(2))
print(ceil(4.7))
print(floor(6.8))
````

````python
from math import *

print(sin(2))
print(cos(3))
print(ceil(6.3))
print(floor(9.9))
````

[as 구문]

모듈을 가져오다 보면 이름의 충돌이 발생하는 경우가 있습니다. 또는 너무 길어 짧게 줄여서 사용하고 싶은 경우도 있을겁니다.

이럴때 모듈의 식별자를 만들어 사용할 수 있는 문법이 as 구문입니다.

````python
import math as m

print(m.sin(1))
print(m.cos(4))
print(m.ceil(7.2))
print(m.floor(8.7))
````

[datetime 모듈]
- 날짜와 시간 관련된 모듈

````python
import datetime

print("오늘 날짜와 시간 출력:")

# 현재 날짜와 시간 가져오기
now = datetime.datetime.now()

# 날짜와 시간 출력
print(f"{now.year} 2025년")
print(f"{now.month} 1월")
print(f"{now.day} 10일")
print(f"{now.hour} 1시")
print(f"{now.minute} 46분")
print(f"{now.second} 30초")

````

````python
import datetime

print("다양한 시간 포맷")

# 현재 날짜와 시간 가져오기
now = datetime.datetime.now()

# 다양한 포맷으로 출력
output_1 = now.strftime("%Y.%m.%d %H:%M:%S")  # 기본 포맷: 년.월.일 시:분:초
output_2 = f'{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초'  # f-string 포맷
output_3 = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")  # .format()을 이용한 포맷

# 출력
print(output_1)
print(output_2)
print(output_3)
print()
````

[time 모듈]
- 시간과 관련된 기능을 다루는 모듈
- 특정 시간동안 코드를 정지할 수 있는 기능이 있음

````python
import time

# 1. 시간 지연 (정지) 기능
print("5초 뒤에 무슨일이 일어날까요?")
time.sleep(3)  # 3초 동안 프로그램을 멈춤
print("5초 뒤에 아무일도 일어나지 않았습니다")

# 2. 코드 실행 시간 측정
start_time = time.time()  # 시작 시간 기록

# 예시로 2초 동안 대기
time.sleep(2)

end_time = time.time()  # 끝 시간 기록
execution_time = end_time - start_time  # 실행 시간 계산
print(f"코드 실행 시간: {execution_time}초")
````

[urllib 모듈]
- URL을 다루는 라이브러리

````python
import urllib.request

# 웹 페이지의 내용을 가져오기
url = "https://www.daum.net/"
response = urllib.request.urlopen(url)

# HTML 내용을 출력
html = response.read()
print(html)
````
